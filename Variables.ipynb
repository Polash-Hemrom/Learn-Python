### Unpacking

```python
data = ['Apple', 2, 4.2]
x, y, z = data
print(data)
```

```python
## Return value Unpacking
def get_values ():
    return (2, 4, 6)

a, b, c = get_values()
print(a)
print(b)
print(c)
```

```python
## *Asterisk unpacking

first, *rest = [2, 4, 6, 8, 10]
print(first)
print(rest)
```

```python
# Mismattched Count

a, *b = [2, 4, 6, 8, 10]    # ValueError
a, *b, c = [2, 4]           # ValueError

print(a)
print(b)
print(a)
print(c)
```

### type() and isinstance()

```python
type(2)         # <class 'int'>
type(4.2)       # <class 'float'>
type('hi')      # <class 'str'>
type([2, 4])    # <class 'list'>
```

### Returning types from unpacked data

```python
data = ['apple', 2, 4.4]
a, b, c  = data
result = (type(a), type(b), type(c))
print(result)
```

### isinstance()

```python
# isinstance(obj, type)

isinstance(2, int)      #True
isinstance(4.2, float)  #True
isinstance('hi', str)   #True
isinstance(2,(int, float))  #True
```

```python
# Subclass Handling

class child(list):
    pass

c = child()
type(c) == list
isinstance(c, list)
```

### Global Variables and the global keyword

```python
x = 2

def set_local():
    x = 4   # Local x: global x unchanged
    print(x)

set_local()
print(x)
```

```python
# The global keyword: Use global name to assign to a global variable from inside a function. Without it, assignment creates a local.

x = y = z =None
data = ['Apple', 2, 4.2]

def assign_globals():
    global x, y, z

assign_globals()
print(assign_globals)
```

```python
# Read vs write: Reading a global does not require global. Only assignment does.

count = 0

def read_global():
    print(count)

def write_global():
    global count
    count += 2
```
