## Match

```python
status = 4
match status:
    case 2:
        label = 'pending'
    case 4:
        label = 'active'
    case 6:
        label = 'done'
    case _:
        label = 'unkown'

print(label)
```

#### Literal Pattern

```python
x = 0
match x:
    case 0:
        print('Zero')
    case None:
        print('None')

match None:
    case None:
        print('None1')
```

```python
# Example with strings:

cmd = 'start'

match cmd:
    case 'start':
        action = 'run'
    case 'stop':
        action = 'halt'
    case 'pause':
        action = 'wait'
    case _:
        action = 'unknown'
```

#### Capture Patterns

```python
data = [2, 4, 6]
match data:
    case [a, b, c]:
        total = a + b + c
        print(total)
```

#### Wildcard Pattern

```python
code = 10
match code:
    case 2:
        result = 'low'
    case 4 | 6:
        result = 'mid'
    case _:
        result = 'other'

print(result)
```

#### | OR Patterns

```python
n = 9

match n:
    case 2 | 4 |6:
        kind = 'even_small'
    case 8 | 10:
        kind = 'even_large'
    case _:
        kind = 'other'

print(kind)
```

```python
key = 'quit'

match key:
    case 'q' | 'quit' | 'exit':
        exit()
    case 's' | 'save':
        do_save()
    case _:
        pass
```

#### AS Patterns

```python
item = [2, 4, 6]

match item:
    case [2, 4, 6] as triple:
        print(triple)
    case(x, y) as pair:
        print(pair)
```

### Sequence Pattern

```python
point = [2, 4]
match point:
    case [0,0]:
        label = 'origin'
    case [x, y]:
        label = f'({x}, {y})'

print(label)
```

```python
items = [2, 4, 6, 8]

match items:
    case [first, *rest]:
        print(first, rest)
```

### Different Length

```python
def describe(seq):
    match seq:
        case []:
            return 'empty'
        case [a]:
            return f'one: {a}'
        case [a, b]:
            return f'two: {a}, {b}'
        case [a, b, *rest]:
            return f'many: {a}, {b}, rest = {rest}'
        
describe([])
describe([4])
describe([2, 4])
describe([2, 4, 6, 8])
```
