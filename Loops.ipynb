## Loops

#### For Loops & range()

```python
for i in range(0, 20, 2):
    print(i)
```

```python
# Loop over a list

pets = ['cat', 'dog', 'hamster', 'parrot']
for pet in pets:
    print(f'I love my {pet}')

print()     # Blank line
```

```python
# Loop over a range of numbers
# range(5) gives : 0, 1, 2, 3, 4

for i in range(5):
    print(f'count: {i}')

print()     # Blank line


# range (start, stop, step)
for i in range (2, 22, 2):
    print(f'count: {i}')
```

## While Loop

#### While Loops (When You Don't Know How many times)

```python
# Countdown Timer

countdown = 5
while countdown > 0:
    print(f'{countdown}')
    countdown -=1
print('Liftoff!🤪 ')



# Guessing Game Logic
import random

secret = random.randint(1, 10)
guess = 0
attempts = 0

while guess != secret:
    guess = random.randint(1, 10)
    attempts += 1


print(f'Found {secret} in {attempts} attempts!')
```

#### Break, Continue & Enumerate

```python
# break - stop when you find what you need

numbers = [4, 7, 2, 9, 1, 5]
for n in numbers:
    if n == 9:
        print(f'Found 9! Stopping.')
        break
    print(f'Checking {n}...')
```

```python
# Continue - skip odd numbers

for i in range(10):
    if i %  2 !=0:
        continue    # skip odd numbers
    print(i, end=' ')
```

```python
# enumerate - get index AND value

fruits = ['Mango', 'Banana', 'Pineapple']
for index, fruit in enumerate(fruits):
    print(f'{index + 1}. {fruit}')
```

```python
# enumerate with custom start

for rank, fruit in enumerate(fruits, start=1):
    print(f'#{rank}: {fruit}')
```

#### Nested Loops

```python
# Multiplication table (3x3)

for row in range(1, 4):
    for col in range(1, 4):
        result = row * col
        print(f'{result:4d}', end = ' ')
    print()


# Finding paris that sum to 10
numbers = [2, 4, 6, 8, 3, 7]
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] + numbers [j] == 10:
            print(f'{numbers[i]} + {numbers[j]} = 10')
```
