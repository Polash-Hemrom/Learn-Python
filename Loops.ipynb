## Loops

#### For Loops & range()

```python
# Loop over a list

pets = ['cat', 'dog', 'hamster', 'parrot']
for pet in pets:
    print(f'I love my {pet}')

print()     # Blank line
```

```python
# Loop over a range of numbers
# range(5) gives : 0, 1, 2, 3, 4

for i in range(5):
    print(f'count: {i}')

print()     # Blank line


# range (start, stop, step)
for i in range (2, 22, 2):
    print(f'count: {i}')
```

## While Loop

#### While Loops (When You Don't Know How many times)

```python
# Countdown Timer

countdown = 5
while countdown > 0:
    print(f'{countdown}')
    countdown -=1
print('Liftoff!🤪 ')



# Guessing Game Logic
import random

secret = random.randint(1, 10)
guess = 0
attempts = 0

while guess != secret:
    guess = random.randint(1, 10)
    attempts += 1


print(f'Found {secret} in {attempts} attempts!')
```
