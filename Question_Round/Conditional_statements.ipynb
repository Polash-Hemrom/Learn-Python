```python
#Questions: Decision Making

# Given two integers, n and m.
# The task is to check the relation between n and m.
# Print "less" if n < m,  "equal" if n == m, and "greater" if n > m.
```

```python
n = int(input())
m = int(input())

if n < m:
    print('less')
elif n == m:
    print('equal')
elif n > m:
    print()
```

```python
# Questions : Switch Statement

# Given a number n, use a switch statement to return "One" if the given
# number is equal to 1, "Two" if the number is 2 and so on till 9 ("Nine")
# else return "Unknown"(without quotes). 
```

```python
n = int(input())

match n:
    case 1:
        print('One')
    case 2:
        print('Two')
    case 3:
        print('Three')
    case 4:
        print('Four')
    case 5:
        print('Five')
    case 6:
        print('Six')
    case 7:
        print('Seven')
    case 8:
        print('Eight')
    case 9:
        print('Nine')
    case _:
        print('Unknown')
```

```python
# Questions : Vowel or Not

# Given a character ch representing an English alphabet
# determine whether it is a vowel or not. Return true if ch is a
# vowel otherwise return false.
```

```python
char = ('a')

if char.lower() in 'aeiou':
    print(f'{char} is vowel')
else:
    print(f'{char} is not a vowel')
```
