### Numeric Types

```python
# int

x = 2
y = -42
z = 10**6

print(x)
print(y)
print(z)
```

```python
# Float

price = 82.6
ratio = 2.0

print(type(price))
print(isinstance(ratio, float))
```

```python
# Complex
# Complex stores numbers with real and imaginary parts.
value = 2 + 4j
value.real
value.imag
```

```python
# Bool
# bool has two values: True and False.

is_ready = True
is_empty = False

# bool is a subclass of int
true_false = isinstance (True, int)
sum = True + True
sub = False + False


print(true_false)
print(sum)
print(sub)
```

### Text Type (str)

```python
# String are immutable Unicode text

name = 'Polash'
lang = 'Python'


# You can index and slice strings, but not mutate characters in place.

s = 'Python'
s[0]    # 'P'
s[1:1]  # 'yth'
```

### Sequence Types

```python
# List (mutable)
# List are ordered, allow duplicates, and can be modified

nums = [2, 4, 6]
nums.append(8)
nums.pop(2)
nums.append (12)

print(nums)
```

```python
r = range(0,8,2)
list(r)
```

### Mapping Type (dict)

```python
# Dictinaries store  key-value pairs

user = {'name' : 'Polash', 'score' : 82}
user['score']
user['score'] = 84

print(user)
```

### Set Types

```python
# Set (mutable)

vals = {2, 4, 4, 6}
print(vals)
```

```python
# Frozenset (immutable)

f = frozenset({2,4,6}) # f.add(8)
print(f)
```

### None Type(None)

```python
result = None

if result is None:
    print('No Result.')
```

##### Immutable: int, float, bool, str, tuple, frozenset, NoneType
##### Mutable: list, dict, set

```python
a = [2, 4]
b = a
b.append(6)

print(a)
print(b)

# a is also [2, 4, 6] because both names reference the same list.
```

```python
# For immutable types, operations produce new objects:

x = 2
y = x

y += 2

# x is still 2; y is 4
```

```python
x = float(5)
print(x)
```

```python
my_dict = {'name' : 'polash', 'age' : 25}
print(type(my_dict))
```

```python
num = -5

if num > 0:
    print('positive')
elif num <= 0:
    print('negative')
else:
    print('zero')
```
