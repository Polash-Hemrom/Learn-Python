### Basic Conditionals

```python
temperature = int(input('Enter temperature: '))


if temperature >= 30:
    print('Its hot! Get more some ice cream')
    print('Dont forget sunscreen!')

elif temperature >= 20:
    print('Nice weather for a walk!')

elif temperature >= 10:
    print('Better grab a jacket!')
else:
    print('Stay inside, its freezing!')
```

```python
## Only ONE brach runs - the first one that's True

print('\n --- Grade Calculator ---')

score = 87

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"


print(f'Score {score} Grade {grade}')
```

### Falsy/Truthy Ternary Expressions

```python
# Using truthiness to check for empty values

username = " "

if username:
    print(f'Welcome , {username}!')
else:
    print('Please enter a username')
```

```python
# Checking a List
inventory = ['sword', 'shield']

if inventory:
    print(f'You have {len(inventory)} items')
else:
    print('Your invetory is empty')
```

```python
# Ternary Expressions (one-linear if/else)

age = 15
status = 'Adult' if age >= 18 else 'minor'
print(f'Age {age} : {status}')


# Ternary in an f-string
coins = 0
print(f'You have {coins} coin{'s' if coins != 1 else ' ' }')
```

### Nested Conditions & Combined Logic

```python
# Amusement park ride checker

age = 5
height = 140    #cm
has_parent = True

# Nested approach (works but gets messy)

if age >= 10:
    if height >= 130:
        print('You can ride')
    else:
        print('Sorry! too short for this ride.')
else:
    print('Sorry! you are too young for this ride')



# Cleaner with 'and/'or'
can_ride_alone = age >= 10 and height >= 130
can_ride_with_parent = age >= 6 and has_parent

if can_ride_alone:
    print('Hop on!')
elif can_ride_with_parent:
    print('You can ride with a parent')
else:
    print('Maybe next year!')
```

```python
# Grade Calculator


score = 85

if score >= 90:
    print('A')
elif score >= 80:
    print('B')
elif score >= 70:
    print('C')
elif score >= 60:
    print('D')
else:
    print('F')
```

```python
# Leap Year Check

year = 2000
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print('Leap Year')
else:
    print('Not a leap year!')
```
